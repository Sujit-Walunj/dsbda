```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('tested.csv')
titanic_data.head(5)
```

```python
titanic_data.info()
```

```python
titanic_data.isna().sum()
```

```python
# Convert 'Survived' column to string if it's numeric
titanic_data['Survived'] = titanic_data['Survived'].astype(str)

# Now create the count plot
# Visualize survival count by sex
sns.countplot(data=titanic_data, x='Sex', hue='Survived')
```

```python
numeric_cols = titanic_data.select_dtypes(include=['float64', 'int64']).columns 
correlation_matrix = titanic_data[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1, linecolor='black')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# data: This parameter specifies the DataFrame (titanic_data in this case) containing the data to be plotted.
# x: This parameter specifies the variable to be plotted on the x-axis of the histogram. In this case, it's 'Fare', which likely represents the ticket fare of the passengers.
# hue: This parameter is optional and is used to differentiate the data by another variable. Here, it's 'Survived', meaning that the histogram bars will be colored differently based on whether the passengers survived or not.
# kde: This parameter is a boolean (True or False) that determines whether to plot a kernel density estimate (KDE) along with the histogram. When set to True, it overlays a smooth line on the histogram to show the probability density function of the data.





sns.histplot(data=titanic_data, x = 'Fare', hue= 'Survived', kde= True);
```
