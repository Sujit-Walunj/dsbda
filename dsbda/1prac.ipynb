```python
region = df['origin'].unique()
# After executing this code, the variable 'region' will hold an array containing
#all unique values found in the 'origin' column of your DataFrame. 
```

```python
print(region)
```

```python
mapping = {'usa':1,'japan':2,'europe':3}
#This line defines a dictionary called 'mapping' where keys are the original region names ('usa', 'japan', 'europe') and values are the corresponding encoded values (1, 2, 3).
df['Origin_encoded'] = df['origin'].map(mapping)
#This line creates a new column 'Origin_encoded' in your DataFrame 'df'.
# It maps the values in the 'origin' column to their corresponding encoded values
# using the 'mapping' dictionary.
```
